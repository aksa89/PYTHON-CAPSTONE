```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/project_data - myexcel.csv.csv')
data
```

# checking for any null values



```python
data.isnull().sum()
```

```python
data.dropna ( inplace = True)
```

```python
data.isnull().sum()
```

# REPLACE THE HEIGHT COLUMN WITH RANDOM VALUES BETWEEN 150 and 180

```python
data['Height'] = np.random.uniform(150,180,size = len(data))
data
```

#Checking for duplicate rows

```python
data.drop_duplicates(inplace = True)
data
```

# 1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.

```python
team_count = data['Team'].value_counts()
team_count
```

# percentage split relative to the total number of employees(round to 2 decimal points)


```python
percent_team_count = team_count/len(data)*100
percent_team_count.round(2)
```

```python
# graphical representation of employees in each team

plt.figure(figsize=(10, 6))
team_count.plot(kind='bar', color='red', edgecolor='black')

# Customize the plot
plt.title("Distribution of Employees Across Teams", fontsize=16)
plt.xlabel("Teams", fontsize=12)
plt.ylabel("Number of Employees", fontsize=12)
plt.xticks(rotation=45,ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

# 2. Segregate employees based on their positions within the company.

```python
employees = data.groupby('Position')['Name'].apply(list)
for Position, Name in employees.items():
    print(f"Employees in {Position} Position:")
    for name in Name:
     print(name)
    print("\n")
```

# 3.Identify the predominant age group among employees.

```python
data['Age Group'] = data['Age'].apply(lambda age:'20-25' if 20 <= age <= 25 else ('26-30' if 26 <= age <= 30 else ('31-35' if 31 <= age <= 35 else 'above 36')))

data
```

```python
age_counts = data['Age Group'].value_counts()
predominant_age_group = age_counts.idxmax()
predominant_count = age_counts.max()

print("Age group distribution:\n", age_counts)
print("\nPredominant age group is", predominant_age_group, "with", predominant_count, "employees")
```

```python
# Plotting age group distribution

plt.figure(figsize=(8, 6))
age_counts.plot(kind='bar', color='green', edgecolor='blue')

plt.title("Age Group Distribution", fontsize=14)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Number of Employees", fontsize=12)
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--')

# Show the plot
plt.tight_layout()
plt.show()
```

# 4. Discover which team and position have the highest salary expenditure

```python
salary_spend = data.groupby(['Team','Position'])['Salary'].sum()
salary_spend.idxmax()
print("\nThe highest salary expenditure is for",salary_spend.idxmax())
```

# 5.Investigate if there's any correlation between age and salary, and represent it visually.

```python
correlation = data['Salary'].corr(data['Age'])
print("THE CORRELATION Between Salary and Age is:",correlation)
```

```python
# visual representation
sns.scatterplot(x="Age" ,y= "Salary",data= data)
plt.ylabel("Salary")
plt.xlabel("Age")
plt.title("Correlation between Salary and Age")
plt.show()
```
